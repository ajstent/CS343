### 2. Plot data inline with other code, modify quickly

```python
# import matplotlib plotting library
import matplotlib.pyplot as plt

# make a line plot: generate x 50 values evenly spaced between -5 and +5
x = np.linspace(-5, 5)

# and associated (random) y values
y = np.random.rand(50)

# Make the plot
plt.plot(x, y)

# Remove the printing of the matplotlib object
plt.show()
```

```python
# Make a straight line and parabola
y_1 = x
y_2 = np.square(x)

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.show()
```

```python
# Adjust the colors of the curves (e.g. black and blue)
plt.plot(x, y_1, 'k')
plt.plot(x, y_2, 'b')
plt.show()
```

```python
# Adjust the line styles of the curves (e.g. dotted/dashed)
plt.plot(x, y_1, 'k--')
plt.plot(x, y_2, 'b.')
plt.show()
```

```python
# Add axes labels, title, legend, x/y limit
plt.plot(x, y_1, 'k--', label='line')
plt.plot(x, y_2, 'b.', label='parabola')
plt.xlabel('My x axis')
plt.ylabel('My x axis')
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.title('My pretty plot!')
plt.legend()
plt.show()
```

```python
# Matplotlib is powerful! Let's create a grid of stacked plots with
# plt.subplots.
# Let's make it 2 rows and 1 col. 
# Let's put the line/parabola plot on top and a scatter plot of random
# data below
#
# The "grid" object of 2 subplots is called a Figure object
# The subplots are called Axes
# Each axis acts like plt above, except the plot decoration commands
# usually have a set_ before them. e.g. ax.set_xlabel('My x label')
# Control figure size with figsize optional parametr of plt.subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5,10))
axes[0].plot(x, y_1, 'k--', label='line')
axes[0].plot(x, y_2, 'b.', label='parabola')
# set the title
axes[0].set_title('Line plot')

# Generate random Gaussian data with mean 5 and stdev 3 in both x and y
x_r = np.random.normal(loc=5, scale=3, size=(30,))
y_r = np.random.normal(loc=5, scale=3, size=(30,))

# Make scatter plot
axes[1].scatter(x_r, y_r)
axes[1].set_title('Scatter plot')
plt.show()
```

```python
# Very useful for this class: if you want to color-code
# scatter plot values, use the c= optional paramerer of plt.scatter
color_inds = np.random.randint(low=0, high=2, size=len(x_r))
plt.scatter(x_r, y_r, c=color_inds)
plt.show()
```
